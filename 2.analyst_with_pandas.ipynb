```python
titanic_data = pd.read_csv('data/Titanic-Dataset.csv')
```

```python
#Bỏ cột
titanic_data.info()
titanic_data.drop(['Cabin'], axis=1, inplace=True )
titanic_data.info()
```

```python
#Bỏ dòng chứa dữ liệu NA
titanic_data.dropna().info()
titanic_data.dropna(subset=['Embarked', 'Age']).info()
titanic_data.fillna({'Age':0, 'Embarked':'Unknown'}).info()
print(titanic_data.Age[:6])
print(titanic_data.Age.fillna(method='ffill')[:6])
print(titanic_data.Age.fillna(method='bfill')[:6]) 
```

Kiểm soat duplicates

```python
#Xóa dòng dữ liệu trùng nhau
df = pd.DataFrame({'A':[1,1,3,4,5,1], 'B':[1,1,3,7,8,1], 'C':[3,1,1,6,7,1]})
df[df.duplicated()]
df[df.duplicated(subset=['A', 'B'])]
```

Hàm ứng dụng

```python
#map chuyển đổi 1 cột thành cột khác
#apply: áp dụng vào 1 series
#grounby: nhóm dữ liệu từng nhóm
titanic_data_age = titanic_data[titanic_data.Age.notnull()]
titanic_data['age_category'] = titanic_data.Age.map(lambda age: 'Kid' if age < 18 else 'Adult')
titanic_data.Age.apply('sum')
titanic_data.Age.apply(lambda age:'Kid' if age < 18 else 'Adult')
def func(e):
    if e.Sex == 'male':
        return e.Fare * 2
    else:
        return e.Fare
titanic_data.apply(func, axis=1)[:5]

titanic_data.groupby(['Sex']).Age.mean()
titanic_data.groupby(['Sex']).Age.agg(['mean','min','max'])

titanic_data[titanic_data.Name.str.contains('Mr')]
```

```python
#Append, Merge, Join và Concatenate
#append: nối các khung dữ liệu
#merge: nối dựa vào series chung
df1 = pd.DataFrame(data=np.random.randint(1,10, size=(10,3)),
columns=list('ABC'))
df2 = pd.DataFrame(data=np.random.randint(1,10,size=(10,3)), columns=list('ABC'))
new=pd.concat([df1,df2])

left = pd.DataFrame({'key':['K0','K1','K2','K3','K4','K5'],
                     'A': ['A0', 'A1', 'A2', 'A3', 'A4','A5'],
                       'B':['B0','B1','B2','B3','B4','B5']})
right = pd.DataFrame({'key': ['K0','K1','K2','K3','K6','K7'],
                       'C':['C0','C1', 'C2', 'C3', 'C6', 'C7'],
                         'D':['D0','D1','D2','D3','D6','D7']})
left.merge(right, on = 'key')
```

```python
#join: ghep du lieu
left = pd.DataFrame({'A':['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']}, index=['K0', 'K1', 'K2'])
right = pd.DataFrame({'C': ['C0', 'C2', 'C3'], 
                      'D': ['D0', 'D2', 'D3']}, index=['K0', 'K2', 'K3'])
print(left.join(right))
```
